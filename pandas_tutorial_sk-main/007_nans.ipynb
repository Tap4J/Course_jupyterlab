## 1. Mazanie stlpcov s NaN hodnotami

```python
# df.dropna()
# df.dropna(how="all") # Mazanie riadkov kde su len NaN hodnoty 
# df.dropna(how="all", axis=1) # To iste pre stlpce
df.dropna(axis=1) # Mazanie stlpca kde je aspon jedna hodnota 
```

```python
# Mazanie riadkov kde su NaN hodnoty na zaklade dvoch stlpcov
target_columns = df.loc[:, ["Alley" , "Fence"]]
row_mask = target_columns.notna().apply(any, axis=1) # maska ktoru aplikujem v locu
df.loc[row_mask, :]
```

```python
# Nazvy ciselnych stlpcov s aspon jednou NaN hodnotou
df.loc[:, (df.dtypes == np.int64) | (df.dtypes == np.float64)].isna().apply(any).index.values
```

## 2. Nahradzovanie NaN hodnot

```python
any(df.GarageYrBlt.isna())
```

```python
df.loc[df.GarageYrBlt.isna(), :]
```

```python
# Takto nahradim NaN hodnotu v stlpci priemerom
df.loc[df.GarageYrBlt.isna(), "GarageYrBlt"] = df.GarageYrBlt.mean()
```

```python
df.GarageYrBlt.mean()
df.GarageYrBlt.median()
df.GarageYrBlt.min()
df.GarageYrBlt.max()
```

```python
A = df.Alley
A.value_counts()
most_freq = A.mode()[0]
A.fillna(most_freq)
A
```

```python
any(df.GarageYrBlt.isna())
```

```python
# Takto nahradim NaN hodnoty najcastejsie sa opakujucou hodnotou
df.loc[df.Alley.isna(), "Alley"] = df.Alley.value_counts().index.values[0]
```
